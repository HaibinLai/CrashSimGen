```python
import matplotlib.pyplot as plt
import numpy as np

# 假设的数据
models = ['DDPM', 'DDIM']

# 训练时长（以分钟为单位）
time_minutes = [89.3 , 67.01]  # DDPM 的时间是 2小时46分钟，DDIM 的时间是 49分钟

# 最终损失值
final_loss = [0.00327464


, 0.002600512
]  #  DDPM , DDIM 

# 设置位置
x = np.arange(len(models))

# 创建图形
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4.5))

# 绘制训练时长条形图
bars1 = ax1.bar(x, time_minutes, color=['blue', 'red'])
ax1.set_title('Training Time Comparison')
ax1.set_ylabel('Time (Minutes)')
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.set_ylim(0, max(time_minutes) + 30)  # 给y轴留一点空间

# 在训练时长条形图上添加数字标签
for i, bar in enumerate(bars1):
    ax1.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 2,  # x, y 坐标
             f'{bar.get_height():.1f}',  # 显示数字，保留1位小数
             ha='center', va='bottom', fontsize=12)

# 绘制最终损失值条形图
bars2 = ax2.bar(x, final_loss, color=['blue', 'red'])
ax2.set_title('Final Loss Comparison on step 4500')
ax2.set_ylabel('Final Loss Value')
ax2.set_xticks(x)
ax2.set_xticklabels(models)
ax2.set_ylim(0, max(final_loss) + 0.001)  # 给y轴留一点空间

# 在最终损失值条形图上添加数字标签
for i, bar in enumerate(bars2):
    ax2.text(bar.get_x() + bar.get_width() / 2, bar.get_height() ,  # x, y 坐标
             f'{bar.get_height():.6f}',  # 显示数字，
             ha='center', va='bottom', fontsize=12)

# 显示图形
plt.tight_layout()
plt.savefig('comparison_on_time.png')
plt.show()
```
