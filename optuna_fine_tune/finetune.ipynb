```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 损失值数据
losses = [0.007229907438158989, 0.005876998417079449]
models = ['Origin Model', 'fine-tuned Model']

# 设置位置
x = np.arange(len(models))

# 创建图形
fig, ax = plt.subplots(figsize=(8, 6))

# 使用seaborn的调色板
sns.set(style="whitegrid")  # 设置背景为网格样式
palette = sns.color_palette("Set2", 2)  # 使用Seaborn的调色板，选择两种颜色

# 绘制条形图
bars = ax.bar(x, losses, color=palette)

# 设置标题和标签
ax.set_title('Loss Comparison Between Origin and fine-tuned Models', fontsize=14)
ax.set_ylabel('Loss Value', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=12)
ax.set_ylim(0.004, 0.008)  # 设置y轴范围

# 在条形图上添加数字标签
for i, bar in enumerate(bars):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.00005,  # x, y 坐标
            f'{bar.get_height():.6f}',  # 显示数字，保留6位小数
            ha='center', va='bottom', fontsize=12)



# 显示图形
plt.tight_layout()

plt.savefig('comparison_on_loss.png')
plt.show()
```
